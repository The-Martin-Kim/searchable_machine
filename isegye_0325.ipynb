```python
song_counts = df['Artist'].value_counts()

plt.pie(song_counts,
        labels=song_counts.index,
        autopct='%1.1f%%')
plt.title('데이터베이스 속 멤버별 곡 수')
plt.axis('equal')
plt.show()
```
